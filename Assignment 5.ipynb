```python
import pandas as pd

# Read the dataset
df = pd.read_csv('churn_modelling.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

# Assuming df is your DataFrame
df = df[(df['Exited'] == 0) | (df['Exited'] == 1)]

# Define features and target
X = df.drop('Exited', axis=1)  # Assuming 'Exited' is the target variable
Y = df['Exited']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Let us create the Neural Network
from keras.models import Sequential
from keras.layers import Dense, Dropout

#Initialize ANN
classifier = Sequential()

#Add input layer and hidden layer
classifier.add(Dense(6, activation = 'relu', input_shape = (X_train.shape[1], )))
classifier.add(Dropout(rate = 0.1))

#Add second layer
classifier.add(Dense(6, activation = 'relu'))
classifier.add(Dropout(rate = 0.1))

#Add output layer
classifier.add(Dense(1, activation = 'sigmoid'))
```

```python
#Let us take a look at our network
classifier.summary()
```

```python
#Optimize the weights
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#Fitting the Neural Network
history = classifier.fit(X_train, y_train, batch_size = 32, epochs = 200, validation_split = 0.1, verbose = 2)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

y_pred = (classifier.predict(X_train) > 0.5).astype(int)
accuracy = accuracy_score(y_train, y_pred)
confusion = confusion_matrix(y_train, y_pred)

print("Accuracy Score:", accuracy)
print("Confusion Matrix:\n", confusion)
```
