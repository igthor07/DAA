```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('bank_marketing.csv')
```

```python
# Drop the "Unnamed: 0" column
df = df.drop("Unnamed: 0", axis=1)

# Drop missing values
df = df.dropna()

# Remove duplicate rows
df = df.drop_duplicates()
```

```python
df
```

```python
# Convert categorical variables into dummy variables
df = pd.get_dummies(df, columns=['job', 'marital','education','default','housing','loan','contact','month','poutcome','deposit'])
```

```python
df
```

```python
 X = df.drop('campaign', axis=1)  # Replace 'target_column' with the actual target variable
 y = df['campaign']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# pip install knn
```

```python
# Standardize features (important for KNN)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build KNN model with k=7
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the KNN model on the test dataset: {:.2f}".format(accuracy))
```
